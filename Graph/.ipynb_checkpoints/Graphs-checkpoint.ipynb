```python
# FOR DIRECTED GRAPHS
#Depth First Search

from collections import defaultdict



class Graph:
    
    
    def __init__(self):
        
        self.graph = defaultdict(list)  # here each key of graph(dictionary) will have value pair as a list
        
        
    def addEdge(self,u,v): # DIRECTED GRAPHS
        
        self.graph[u].append(v)
        
    def DFSUtil(self, u, visited):
        visited[u] = True
        print(u, end=" ")
        
        for i in self.graph[u]:
            if visited[i] == False :
                self.DFSUtil(i,visited)  #  notice that while calling the func self is not given as an argument to the function 
                
                
    def DFS(self,u):
        
        visited = [False] * (max(self.graph) + 1)#the max function gives us largest of the input values ,, in case of dictionary it gives us the maximum key
            
        self.DFSUtil(u,visited)
        
    def BFS(self, u):
        visited = [False] * (max(self.graph) + 1)
        tobe_explored = []
        self.BFSUtil(u, visited, tobe_explored)
        
    def BFSUtil(self,u,visited, tobe_explored):
        visited[u] = True
        print(u, end = " ")
        tobe_explored.append(u)
        while len(tobe_explored) != 0 :
            
            u = tobe_explored.pop(0)
            
            for i in self.graph[u]:
                if visited[i] == False:
                    visited[i] = True
                    print(i, end=" ")
                    tobe_explored.append(i)
            
        
        
#Driver code
g = Graph() 
g.addEdge(0, 1) 
g.addEdge(0, 2) 
g.addEdge(1, 2) 
g.addEdge(2, 0) 
g.addEdge(2, 3) 
g.addEdge(3, 3) 

print("Following is DFS from (starting from vertex 2)") 
g.DFS(2) 
print("\nBFS is")
g.BFS(2)
    
```

```python
# reference Udemy(Abdul Bari)
# MST will be constructed started from u
#Complexity O(V**2)

import sys
from functools import reduce
class Graph_Matrix:  #using Adjacency Matrix   #Undirected Graph
    def __init__(self, vertices):
        self.V = vertices
        self.Matrix = [[sys.maxsize for col in range(vertices+1)] for row in range(vertices+1)]
        # you can also make nested dictionary if your vertices are not integers
    
    def addEdge(self,u,v,w):
        self.Matrix[u][v] = w
        self.Matrix[v][u] = w
        
    def Prims_MST(self, u): 
        near  = {}
        for i in range(1, self.V + 1):
            near[i] = -1
        result = [[0,0] for x in range(self.V - 1)]
        
        # First we will find the minimum Edge connected to v
        v = -1  
        m = sys.maxsize
        for i in range(self.V):
            if self.Matrix[u][i] < m:
                m = self.Matrix[u][i]
                v = i
                
        near[u] = 0
        near[v] = 0

        result[0][0] = u
        result[0][1] = v
        
        # now we will update whole 'near' array
        for i in range(1,self.V + 1):
            if near[i] != 0:
                if self.Matrix[u][i] < self.Matrix[v][i]:
                    near[i] = u
                else:
                    near[i] = v
        print(near)
                
        for i in range(self.V - 2):
            m = sys.maxsize
            k = 0
            for j in range(1,self.V + 1):
               # print(self.Matrix[j][near[j]])
                if near[j] != 0 and self.Matrix[j][near[j]] < m:
                    m = self.Matrix[j][near[j]]
                    k = j
            #print(k)
            result[i+1][0] = k
            result[i+1][1] = near[k]
            
            near[k] = 0 # Edge now has been included
            
            #update 'near' array
            for j in range(1,self.V+1):
                if near[j] !=0 and self.Matrix[j][near[j]] > self.Matrix[j][k] :
                    near[j] = k
        return (result, reduce(lambda a,b: a+b, [self.Matrix[i][j] for i,j in result]))
            
g1 = Graph_Matrix(7)
g1.addEdge(1, 6, 6)
g1.addEdge(1, 2, 25)
g1.addEdge(6, 5, 20)
g1.addEdge(5, 7, 18)
g1.addEdge(2, 7, 10)
g1.addEdge(4, 5, 16)
g1.addEdge(7, 4, 14)
g1.addEdge(3, 4, 8)
g1.addEdge(3, 2, 12)


        
```

```python
# refer coding ninjas video to understand this algo
# Complexity remains the same as before when we used matrix that is O(V**2)
from collections import defaultdict
import sys

class WeightedGraph:
    def __init__(self):
        self.graph = defaultdict(lambda: [])  # here each key of graph(dictionary) will have value pair as a list

    def addEdge(self, u, v, w): 
        """Adds Edge according to sorted weights
           Right now, sorting weights doesnt have any effect on time complexity
           I am doing this just for fun"""
        
        # storing u to v edge in sorted weight order
        index = len(self.graph[u])
        for i in range(len(self.graph[u])):
            if self.graph[u][i][1] > w:
                index = i
                break
                
        if index == len(self.graph[u]):
            self.graph[u].append((v, w))
        else:
            self.graph[u] = self.graph[u][:index] + [(v, w)] + self.graph[u][index:]
            
        # storing v to u edge in sorted weight order
        index = len(self.graph[v])
        for i in range(len(self.graph[v])):
            if self.graph[v][i][1] > w:
                index = i
                break

        if index == len(self.graph[v]):
            self.graph[v].append((u, w))
        else:
            self.graph[v] = self.graph[v][:index] + [(u, w)] + self.graph[v][index:]

    def isEmpty(self):  # return True if Empty else False
        if (self.vertex_count() == 0):
            return True
        else:  # if the vertices are there but there are no edges then also the graph is empty
            return all([len(list1) == 0 for list1 in self.graph.values()])

    def vertex_count(self):
        return len(self.graph)

    def __str__(self):
        return str(dict(self.graph))

    def del_edge(self, u, v, w):
        for i in range(
                len(self.graph[u])):  # v to u will not be the first in the list , so we need to search u in the list[v]
            tup = self.graph[u][i]
            if tup[0] == v and tup[1] == w:
                self.graph[u].pop(i)
                break
        for i in range(
                len(self.graph[v])):  # v to u will not be the first in the list , so we need to search u in the list[v]
            tup = self.graph[v][i]
            if tup[0] == u and tup[1] == w:
                self.graph[v].pop(i)
                break
        

    def prims_mst(self, source):  # Applicable only for undirected Graphs, Complexity = O(V**2)
        # refer coding ninjas video for understanding the algorithm
        """
        1. repeat this until all the vertex are included in MST:
        -     - find lowest weighted edge in mst_graph whose vertex is not included MST(At very first selected vertex will be source vertex)
        -     - include that vertex in MST
        -     - find adjacent vertices(which are not in MST) of that vertex and store it into mst_graph
        -    loop ends   """ 
        mstSet = {} 
        mst_graph = {} # stores key:value pair such that --- vertex : [parent, weight]
        
        for key in self.graph.keys():
            mstSet[key] = False  # declare all the elements of mset as False as no vertex has been visited yet
            mst_graph[key] = [-1, sys.maxsize]  # declare all the elements of mst_graph

        mst_graph[source] = [source, 0]  # source vertex is its own parent therefore weight = 0

        while all(mstSet.values()) != True:  # runs until all the vertex are not included in the MST
            temp = [-1, sys.maxsize]
            vertex = 0
        
            for key, lis in mst_graph.items(): # select a vertex whose weight is minimium and which is not included in MST
                if mstSet[key] == False: 
                    if lis[1] < temp[1]:
                        temp = lis
                        vertex = key

            mstSet[vertex] = True  # selected minimum weighted vertex will be included in MST 

            for tup in self.graph[vertex]:  # weight and parent of vertices which are adjacent to the selected vertex are being updated 
                if mstSet[tup[0]] == False: # ADJACENT VERTEX SHOULD NOT BE IN MST
                    if tup[1] < mst_graph[tup[0]][1]:  
                        mst_graph[tup[0]] = [vertex, tup[1]]
        return mst_graph  # returns a dictionary
    
    def kruskalMst(self):
        # Refer Abdul Bari
        # First store all the edges in an array
        # Create Disjoint set S, to detect cycles #
        # Result will be stored in another array
        
        
        # storing all the edges into array
        n = len(self.graph)
        S = {}
        for key in self.graph.keys(): # for all the vertices initiate S with -1 because right now they are there own set's parent( refer Disjoint Sets if you didnt get it)
            S[key] = -1
            
        Edges = {}
        for key in self.graph.keys():
            for E in self.graph[key]:
                Edges.append([key, E[0], E[1]])
    
        # all the edges in the graph are stored in array in a sorted manner 
        # which causes complexity to be O(ElogE)
        # Selecting Minimum edge will be O(1) as Edges array will be already sorted
        
        """
        Loop Begins until V-1 vertices are not included in 
        
        """
        
        
        

```
